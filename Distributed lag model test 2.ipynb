```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_excel('/Users/Suwani/Desktop/RMOODYS.xlsx')
```

```python
df.head()
```

```python
X = df[['AGG_MEAN_comp','LAG_1']]
Y = df[['Change %']]
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
```

```python
est = sm.OLS(Y,X).fit()
```
