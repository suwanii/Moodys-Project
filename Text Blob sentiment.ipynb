```python
import pandas as pd
```

```python
df = pd.read_excel('/Users/Suwani/Desktop/Moodys Project/Cleaned data/To analyze/Labelled data .xlsx')
```

```python
df.head()
```

```python
from textblob import TextBlob

# compute sentiment scores (polarity) and labels
sentiment_scores_tb = [round(TextBlob(article).sentiment.polarity, 3) for article in df['Headline']]
sentiment_category_tb = ['1' if score > 0 
                             else '-1' if score < 0 
                                 else '0' 
                                     for score in sentiment_scores_tb]
```

```python
# sentiment statistics per news category
df_1 = pd.DataFrame([list(df['Headline']), sentiment_scores_tb, sentiment_category_tb]).T
df_1.columns = ['Headline', 'sentiment_score', 'sentiment_category']
df_1['sentiment_score'] = df_1.sentiment_score.astype('float')
#df.groupby(by=['news_category']).describe()
#view raw
```
