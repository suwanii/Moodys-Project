### cleaning text data

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np
np.random.seed(400)
import nltk
from nltk import PorterStemmer
#PorterStemmer().stem('complications')
```

```python
rdf = pd.read_excel('/Users/Suwani/Desktop/Moodys Project/Reuters_cleaned.xlsx')
```

```python
rdf.Headline.head()
```

```python
#Lower case
rdf['Headline_l'] = rdf['Headline'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
rdf.Headline_l.head()
```

```python
#Removing Punctuation
rdf['Headline_l'] = rdf['Headline_l'].str.replace('[^\w\s]','')
```

```python
rdf.Headline_l.head()
```

```python
#Removal of Stop Words
from nltk.corpus import stopwords
stop = stopwords.words('english')
rdf['Headline_l'] = rdf['Headline_l'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
rdf.Headline_l.head()
```

```python
#Rare words removal
freq = pd.Series(' '.join(rdf['Headline_l']).split()).value_counts()[-10:]
```

```python
freq
```

```python
#function to perform the pre processing steps on the entire dataset

def lemmatize_stemming(text):
    return PorterStemmer().stem(WordNetLemmatizer().lemmatize(text, pos='v'))

# Tokenize and lemmatize
def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text) :
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
            
    return result
```

```python
processed_docs = []

for doc in rdf.Headline_l:
    processed_docs.append(preprocess(doc))
```

```python
processed_docs
```

### Bag of words on the dataset

```python
'''
Create a dictionary from 'processed_docs' containing the number of times a word appears 
in the training set using gensim.corpora.Dictionary and call it 'dictionary'
'''
dictionary = gensim.corpora.Dictionary(processed_docs)
```

```python
'''
OPTIONAL STEP
Remove very rare and very common words:

- words appearing less than 15 times
- words appearing in more than 10% of all documents
'''
dictionary.filter_extremes(no_below=15, no_above=0.1, keep_n= 100000)
```

```python
'''
Create the Bag-of-words model for each document i.e for each document we create a dictionary reporting how many
words and how many times those words appear. Save this to 'bow_corpus'
'''
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
'''
Preview BOW for our sample preprocessed document
'''
document_num = 3
bow_doc_x = bow_corpus[document_num]

for i in range(len(bow_doc_x)):
    print("Word {} (\"{}\") appears {} time.".format(bow_doc_x[i][0], 
                                                     dictionary[bow_doc_x[i][0]], 
                                                     bow_doc_x[i][1]))
```

```python
from __future__ import print_function
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 10)
dtm_tf = tf_vectorizer.fit_transform(rdf['Headline_l'])
print(rdf['Headline_l'].shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(rdf['Headline_l'])
print(dtm_tfidf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=10, random_state=0)
lda_tf.fit(dtm_tf)

# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=10, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer)
```
